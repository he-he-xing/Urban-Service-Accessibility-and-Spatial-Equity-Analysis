```python
import requests
import time
```

```python
api_key='*********************'
url='https://restapi.amap.com/v3/place/text'
```

```python
poi_types={
    '综合医院':'090100',
    '专科医院':'090200',
    '诊所':'090300',
    '急救中心':'090400',
    '博物馆':'140100',
    '展览馆':'140200',
    '会展中心':'140300',
    '美术馆':'140400',
     '图书馆':'140500',
    '科技馆':'140600',
    '天文馆':'140700',
    '文化宫':'140800',
    '档案馆':'140900',
    '档案馆':'140900',
    '学校':'141200',
    '公共设施':'200000',
}
```

```python
all_pois = []
for type_name, type_code in poi_types.items():
    page = 1
    while True:
        params = {
            'key': api_key,
            'city': '广州',           # 城市名称
            'citylimit': 'true',       # 强制限制在城市内
            'types': type_code,        # POI类型
            'offset': 20,              # 每页数量
            'page': page,
            'output': 'json'
        }
        
        response = requests.get(url, params=params).json()
        
        if response.get('status') == '1':
            pois = response.get('pois', [])
            all_pois.extend(pois)
            print(f"{type_name}: 第{page}页, 获取{len(pois)}条")
            
            if len(pois) < 20:
                break
            page += 1
        else:
            print(f'请求失败:{response.get("info")}')
            break
        
        time.sleep(0.5)  # 控制请求频率
    
    time.sleep(0.5)

print(f"总共获取 {len(all_pois)} 条POI数据")
```

```python
import pandas as pd
```

```python
all_pois = pd.DataFrame(all_pois).drop_duplicates('id')
```
