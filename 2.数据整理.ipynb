```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from rasterstats import zonal_stats
```

```python
gdf_road=gpd.read_file('road.geojson')
```

```python
pois_public_service=pd.read_csv('pois_public_service.csv')
```

```python
pois_public_service
```

```python
location=pois_public_service['location'].str.split(',',expand=True)
```

```python
location.rename(columns={0:'longitude'},inplace=True)
location.rename(columns={1:'latitude'},inplace=True)
```

```python
pois_public_service=pois_public_service.join(location,how='outer')
```

```python
pois_public_service
```

```python
crs_proj = "EPSG:32650"  
crs_wgs84 = "EPSG:4326"
```

```python
from shapely.geometry import Point
```

```python
geometry=[Point(xy)for xy in zip(pois_public_service['longitude'],pois_public_service['latitude'])]
```

```python
gdf_pois_public_service=gpd.GeoDataFrame(pois_public_service,geometry=geometry,crs=crs_wgs84)
```

```python
gdf_pois_public_service=gdf_pois_public_service[['id','name','typecode','adname','distance','geometry']]
```
